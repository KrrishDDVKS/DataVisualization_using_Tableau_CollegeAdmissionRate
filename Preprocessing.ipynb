```python
df=pd.read_csv("MERGED2018-19.csv")
df=df.dropna(subset=['ADM_RATE'])
x=df['ADM_RATE']
y=df['LATITUDE']
z=df['LONGITUDE']
df.head()
```

```python
df=df.drop(columns='UNITID')
df=df.drop(columns='OPEID')
df=df.drop(columns='OPEID6')
df=df.drop(columns='INSTURL')
df=df.drop(columns='NPCURL')
df=df.drop(columns='SCH_DEG')
df=df.drop(columns='LOCALE2')
df=df.drop(columns='RELAFFIL')
df=df.drop(columns='ADM_RATE_ALL')
df=df.drop(columns='SATVR25')
df=df.drop(columns='SATMT25')
df=df.drop(columns='SATWR25')
df=df.drop(columns='SATVRMID')
df=df.drop(columns='SATMTMID')
df=df.drop(columns='SATWRMID')
df=df.drop(columns='ACTCM25')
df=df.iloc[ :, 0:14]
df['LOCALE']=df['LOCALE'].replace(np.nan,0)
df['LOCALE']=df['LOCALE'].replace(-3,0)
df['ADM_RATE']=x
df
```

```python
def f(t,S,T):
    a=[]
    b=[]
    for i in df[t]:
        j=int(i//10)
        b.append(j)
        k=int(i%10)
        if j>2:
            a.append(S[k])
        else:
            a.append(T[k])
    df['TYPE']=a
    df['PLACE']=b
```

```python
def fn(t,S):
    a=[]
    for i in df[t]:
        a.append(S[i])
    df[t]=a
```
